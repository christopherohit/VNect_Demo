```python
import caffe
import numpy as np
```

```python
# Reference: https://zhuanlan.zhihu.com/p/27298134

def load_net(proto_path, weights_path):
    """Load caffe model"""
    caffe.set_mode_cpu()
    return caffe.Net(proto_path, weights_path, caffe.TEST)


def print_layer_info(net):
    """Print the names and the output shapes of the layers"""
    for layer_name, blob in net.blobs.items():
        print('The output shape of layer {} is {}'.format(
            layer_name, blob.data.shape[1:]))
```

```python
net = load_net('./vnect_net.prototxt', 'vnect_model.caffemodel')
print_layer_info(net)
```

```python
num = 0
for name, param in net.params.items():
    print(name, len(param))
    num += len(param)
    
print('\ntotal num of parameters:', np.sum(num))
```

```python
for name, param in net.params.items():
    print(name)
    for i in range(len(param)):
        print('\t', param[i].data.shape)
```
